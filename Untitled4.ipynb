```python
class Player:
    """Represents one Battleship player."""
    position: int = 0
    name: str = "Nomathemba"
    board: 'Board' = None
    guess_board: 'Board' = None
    fleet: 'Fleet' = None
    is_active: bool = False
```

```python
class Game:
    """Controls the overall Battleship game session."""
    timer: int = 0
    current_turn: int = 0
    players: list = []
    description: str = "Welcome to Battleship!"
    game_over: bool = False
    rules: 'Rules' = None
    scoreboard: 'Scoreboard' = None
    turn: 'Turn' = None
    validator: 'Validator' = None
```

```python
class Board:
    """Represents the 10 by 10 Battleship grid."""
    ships: list = []
    size: int = 10
    shots: list = []
    grid_cells: list = []
```

```python
class Ship:
    """Represents one ship placed on the board."""
    name: str = "Destroyer"
    size: int = 3
    coordinates: list = []
    hits: int = 0
    is_sunk: bool = False
```

```python
class Coordinate:
    """Represents one row and column location on the board."""
    row: int = 0
    col: int = 0
```

```python
class Fleet:
    """Collection of ships owned by a player."""
    ships: list = []
    ships_remaining: int = 5
    owner: 'Player' = None
```

```python
class Rules:
    """Stores game rules and settings."""
    board_size: int = 10
    num_ships: int = 5
    allow_adjacent_ships: bool = False
    max_players: int = 2
```

```python
class Scoreboard:
    """Tracks hits, misses, turns, and winner."""
    hits: int = 0
    misses: int = 0
    turns_taken: int = 0
    winner: str = ""
```

```python
class Shot:
    """Represents a player's attack on a board location."""
    target: 'Coordinate' = None
    shooter: 'Player' = None
    result: str = "Pending"
```

```python
class Turn:
    """Tracks whose turn it is and what shot was taken."""
    current_player: 'Player' = None
    opponent: 'Player' = None
    shot: 'Shot' = None
    turn_number: int = 0
```

```python
class Validator:
    """Checks whether moves and ship placements are allowed."""
    rules: 'Rules' = None
    valid_input: bool = True
    error_message: str = ""
```

```python
g = Game()

g.rules = Rules()
g.scoreboard = Scoreboard()
g.turn = Turn()
g.validator = Validator()

player_1 = Player()
player_1.name = "Nomathewmba"
player_1.position = 1
player_1.board = Board()
player_1.guess_board = Board()
player_1.fleet = Fleet()
player_1.fleet.owner = player_1

player_2 = Player()
player_2.name = "Ted"
player_2.position = 2
player_2.board = Board()
player_2.guess_board = Board()
player_2.fleet = Fleet()
player_2.fleet.owner = player_2

ship_1 = Ship()
ship_1.name = "Destroyer"
ship_1.size = 3
ship_1.coordinates = [Coordinate(), Coordinate(), Coordinate()]
player_1.fleet.ships.append(ship_1)
player_1.board.ships.append(ship_1)

shot_1 = Shot()
shot_1.shooter = player_1
shot_1.target = Coordinate()
shot_1.target.row = 2
shot_1.target.col = 4
shot_1.result = "Pending"

g.turn.current_player = player_1
g.turn.opponent = player_2
g.turn.shot = shot_1
g.players = [player_1, player_2]

print(f"{g.description} The current game time elapsed is {g.timer}.")
print("Players in this game:")

for p in g.players:
    print(f" {p.name}")

print("\nClass relationships created:")
print(f"{player_1.name} owns a fleet with {len(player_1.fleet.ships)} ship.")
print(f"{player_1.name}'s board contains {len(player_1.board.ships)} ship.")
print(f"The current turn belongs to {g.turn.current_player.name}.")
print(f"The shot target is row {g.turn.shot.target.row}, column {g.turn.shot.target.col}.")
```
